## Building n-Butane (UA)

The first task is to build a n-Butane molecule. TraPPE-UA is a united atom (UA) force field that is parameterized for phase equilibria and transferability. The tricky part about Cassandra and TraPPE is that bonds are fixed, and you **must** ensure that the bond distances match that in the force field. Otherwise, you are not using the TraPPE force field properly and Cassandra will throw an error. We will use a tool called "obabel" to help with this.

```python
%set_env SHELL=/bin/bash
%set_env OMP_NUM_THREADS=8
```

```python
%%bash
obabel -:"CCCC" --gen3d -O butane.gzmat
more butane.gzmat
```

With obabel, we created a Gaussian Z-matrix file for UA butane.

To decide what values we need to set for each variable, we need to consult the [TraPPE ForceField](http://trappe.oit.umn.edu/)

```python
with open("butane.gzmat", "w") as fh:
    fh.write("""
!Put Keywords Here, check Charge and Multiplicity.
#

 

0  1
C
C  1  r2
C  2  r3  1  a3
C  3  r4  2  a4  1  d4
Variables:
r2= 1.54
r3= 1.54
a3= 114.0
r4= 1.54
a4= 114.0
d4= 180.0
    """)
```

```python
%%bash
obabel butane.gzmat -O butane.pdb
more butane.pdb
```

Last thing we have to do to get this PDB file ready for Cassandra is append the atom type of each atom in the system as a tag.

```python
with open("species1.pdb", "w") as fh:
    fh.write("""
COMPND    butane.gzmat 
AUTHOR    GENERATED BY OPEN BABEL 3.1.0
HETATM    1  C   UNL     1       0.000   0.000   0.000  1.00  0.00          C3
HETATM    2  C   UNL     1       1.540   0.000   0.000  1.00  0.00          C2
HETATM    3  C   UNL     1       2.166   0.000   1.407  1.00  0.00          C2
HETATM    4  C   UNL     1       3.706  -0.000   1.407  1.00  0.00          C3
CONECT    1    2                                                      
CONECT    2    1    3                                                 
CONECT    3    2    4                                                 
CONECT    4    3                                                      
MASTER        0    0    0    0    0    0    0    0    4    0    4    0
END
    """)
```

# Creating the MCF File

Molecular connectivity files (MCF) contain the topology information for Cassandra per species. Cassandra comes shipped with two python scripts, mcfgen.py and library_setup.py, to assist with setting up cassandra simulations. We'll get to library_setup.py in a minute, for now we will use the mcfgen.py script to set up the Force Field Template and MCF file for butane.

```python
!printf "LJ\nOPLS" | mcfgen.py species1.pdb --ffTemplate
```

```python
import numpy as np
from scipy.constants import Boltzmann, Avogadro

a0 = 0.00
a1 = 355.03
a2 = -68.19
a3 = 791.32
ac = np.array([a0, a1, a2, a3])
print(ac * Boltzmann * Avogadro / 1000.0)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

def read_cassandra_prp(filename):
    with open(filename) as fh:
        #First line
        fh.readline()
        #Secondline
        column_names = fh.readline().split()
        del column_names[0]
    
    return pd.read_table(filename, skiprows=3, names=column_names, delim_whitespace=True)

df_box1_eq = read_cassandra_prp("equil.out.box1.prp")
df_box2_eq = read_cassandra_prp("equil.out.box2.prp")

df_box1_pd = read_cassandra_prp("prod.out.box1.prp")
df_box2_pd = read_cassandra_prp("prod.out.box2.prp")
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2,sharey=True)

ax[0].set_ylabel("Mass Density (kg  m$^{-3}$)")

df_box1_eq.plot(x="MC_SWEEP", y="Mass_Density", ax=ax[0], label="Box 1")
df_box2_eq.plot(x="MC_SWEEP", y="Mass_Density", ax=ax[0], label="Box 2")
df_box1_pd.plot(x="MC_SWEEP", y="Mass_Density", ax=ax[1], label="Box 1")
df_box2_pd.plot(x="MC_SWEEP", y="Mass_Density", ax=ax[1], label="Box 2")
```
