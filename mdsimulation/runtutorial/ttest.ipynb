```python
import numpy
import scipy
import matplotlib.pyplot as plt

def autocorr1D(array):
    '''Takes in a linear numpy array, performs autocorrelation
       function and returns normalized array with half the length
       of the input'''
    ft = numpy.fft.rfft(array-numpy.average(array))
    acorr = numpy.fft.irfft(ft*numpy.conjugate(ft))/(len(array)*numpy.var(array))
    return acorr[0:int(len(acorr)/2)]
data = numpy.loadtxt("e1.txt")
```

```python
plt.plot(data[:,1])
```

```python
start = 600
cut = 700
stop = 800
#Higher T-statistic means more difference between samples
#Lower P-value indicates a low probability that the difference in means was by chance
#P-value close to 1 is a great measure that the two samples have the same stable average
scipy.stats.ttest_ind(data[start:cut,1],data[cut:stop,1],equal_var=True)


```

```python
start = 0
cut = 150
stop = 900
scipy.stats.ttest_ind(data[start:cut,1],data[cut:stop,1],equal_var=True)
```
